```python
import pandas as pd
import numpy as np
data = {"nombre": ["Ana", "Juan", "Pedro", "Maria", "Luis", None],"edad": [23, 35, None, 29, 41, 30],"ciudad": ["CDMX", "cdmx", "Monterrey ", "GDL", "  gdl", "CDMX"],"ingresos": ["$12,000", "$18,500", "$15,300", None, "$22000.", "$17800."]}
```

Manipulaci贸n de Datos


```python
# nota: para importar un .csv se usa el:
# variable = pd.read_csv('NombreArchivo.csv')
df = pd.DataFrame(data)
df.head()
```

```python
# To avoid TypeError when describing mixed-type columns, especially those with non-numeric characters
# that Pandas might internally try to sort numerically, we should clean them.

# First, let's clean the 'ingresos' column by removing non-numeric characters and converting to a numeric type.
# Ensure 'ingresos' column is string type to apply string operations, handling None values.
df['ingresos'] = df['ingresos'].astype(str)

# Remove '$' and ','
df['ingresos'] = df['ingresos'].str.replace('$', '', regex=False)
df['ingresos'] = df['ingresos'].str.replace(',', '', regex=False)

# Strip any leading/trailing whitespace
df['ingresos'] = df['ingresos'].str.strip()

# Replace string 'None' (which results from None in original data + astype(str)) with actual NaN
df['ingresos'] = df['ingresos'].replace('None', np.nan)

# Convert to numeric. 'errors='coerce' will turn any values that cannot be converted into NaN.
df['ingresos'] = pd.to_numeric(df['ingresos'], errors='coerce')

# Now, describe the DataFrame again.
df.describe(include='all')
```

# Introduccion a la regresi贸n lineal

```python
from sklearn.linear_model import LinearRegression
```

Que es una regresi贸n lineal?
Una regresion itenta predecir un valor numerico apartir de otros datos.


Ejemplo:
* Predecir ingresos segun la edad

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df_clean, x='edad', y='ingresos')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.show()
```

```python
# MATRIZ DE CORRELACIN
# 0,0 SIN RELACION
# 0,3 CORRELACION DEBIL
# 0,5 A 0,7 MODERADA
#0,8 O MAS ES FUERTE

df_clean[['edad','ingresos']].corr()
```

```python
# Before fitting the model, handle any NaN values in the target variable Y.
# We will use the mean imputation as done previously for 'edad'.
df_clean['ingresos'] = df_clean['ingresos'].fillna(df_clean['ingresos'].mean())

X = df_clean[['edad']]
Y = df_clean['ingresos']

modelo = LinearRegression()
modelo.fit(X,Y)
```

Tiene relaci贸nc con la forma de la pendiente
Que hace fit? Busca los valores de m, y b que mejor se ajustan a tus datos. Minimiza el error entre los ingresos reales y los predichos


```python
edad_nueva = pd.DataFrame({'edad': [30]})
predicci贸n = modelo.predict(edad_nueva)
predicci贸n
```

```python
# podemos evaluar el modelo con
from sklearn.metrics import mean_squared_error, r2_score
y_pred = modelo.predict(X)
mse = mean_squared_error(Y, y_pred)
r2 = r2_score(Y, y_pred)
#0.3 a 0,5 aceptable
# 0.5 a 0,7 bueno
#0,7 hacia arriba excellent
#ingresos = 尾0 + 尾1 路 edad#尾0  intercepto#尾1  coeficiente de edad
print(r2)
```
