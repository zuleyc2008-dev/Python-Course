#ELIF


```python
# AHORA HAGAMOS UN PROGRAMA EN EL QUE PREGUNTAMOS LA CALFICIACION DEL CERO AL CIEN Y QUEREMOS SABER SU EQUIVALENTE EN EL SISTEMA DE EVALUACIÓN ESTADOUNIDENSE PARA SIMPLICIDAD BASEMONOS EN ESTO.
# A+ = 100
# A = 90 - 99
# B = 80 - 89
# C = 70 - 79
# D = 60 - 69
# F = < 60

cal = 89
if cal == 100:
    print('A+')
elif cal >= 90:
    print('A')
elif cal >= 80:
    print('B')
elif cal >= 70:
    print('C')
elif cal >= 60:
    print('D')
else:
    print('F')
```

# OPERADORES LÓGICOS CON CONDICIONALES

```python
# and, or, not

# Quieres un sandwich con todo
pan = True
jamón = True
Queso = False
print(pan and jamón and Queso)
if not (pan and jamón):
    print('No puedes hacer el sandwich')
```

```python
# AHORA HAGAMOS UN PROGRAMA EN EL QUE PREGUNTAMOS LA CALFICIACION DEL CERO AL CIEN Y QUEREMOS SABER SU EQUIVALENTE EN EL SISTEMA DE EVALUACIÓN ESTADOUNIDENSE PARA SIMPLICIDAD BASEMONOS EN ESTO.
# A+ = 100
# A = 90 - 99
# B = 80 - 89
# C = 70 - 79
# D = 60 - 69
# F = < 60

cal = 0
if cal >= 0 and cal <= 100:
    if cal == 100:
        print('A+')
    elif cal >= 90:
        print('A')
    elif cal >= 80:
        print('B')
    elif cal >= 70:
        print('C')
    elif cal >= 60:
        print('D')
    else:
        print('F')
```

```python
# el modulo (%)
# El modulo es un operador matemático que devuelve el residuo de una división
# 10 % 3 da 1, por que 10 dividido entre tres cabe 3 veces y sobra 1

num = 9
if num % 3 == 0 or num % 7 == 0:
    print('No tiene residuo')

# saber si el número es impar o par
num = 6
if num % 2 == 0:
    print('Par')
else:
    print('Impar')
```

# FOR (ciclo)
Los llamados ciclos for nos ayudan a realizar acciones durante una cantidad determinada de veces


```python
# Imprimir los números del 0 al 4
for i in range(5): #Toma 5 como el número final pero inicia siempre en 0
    print(i)
```

```python
for num in range(2,11):
    print(num)
```

```python
cadena = 'Python'
for letra_por_letra in cadena:
    print(letra_por_letra)
```

```python
# También existe el range(inicio, fin,paso)
for i in range(2,21,2):
    print(i)
```

#WHILE


```python
suma = 0
numeros = int(input('Ingresa el número para sumar:'))

while numeros != 0:
  suma += numeros
  numeros = int(input('Ingresa el numero para sumar:'))
print(f'Tu suma es {suma}')
```

```python
# Usar while como for

numero = 1 # Initialize 'numero' before using it in the loop
while numero <= 12:
    print(numero)
    numero += 1
```
