```python
# hagamos una contraseña aleatoria de lngitud 12
import random
import string
Longitud = 12
caracteres = string.ascii_letters + string.digits + string.punctuation
contraseña = ''.join(random.choice(caracteres) for _ in range(Longitud))

print(contraseña)
```

# JSON ES UN FORMATO PARA GUARDAR Y TRANSMITIR DATOS. ES MUY COMUN EN API's, ARCHIVOS DE CONFIGURACIÓN Y ALMACENAMIENTO DE INFORMACIÓN.

(Se parece a un diccionariod e Python, pero es solo texto)

Para que sirve?

* Sirve para convertir objetos de Python a texto y tambien de texto a objetos de Python


Esto se llama
SERIALIZAR : Convertir un objeto de Python a texto Json
DESSERIALIZAR: Convertir un texto de Json a objeto de Python


```python
# Función principal:

#1) json.dumps(obj): Convierte un diccionario u objeto de Python a sting de Json

import json
obj = {'nombre': 'Ana',
 'edad': 25}
s = json.dumps(obj, ensure_ascii=False)
print(s)
```

```python
#2) json.loads(texto_json): Convierte un string Json en un diccionario Python
restaurar = json.loads(s)
print(restaurar)
# Al venir de un lugar o propetiario desconocido debes de tener cuidado puede tener datos inesperados, informacion incompleta, etc.
```

#NUMPY

Provee el tipo de ndarray, operaciones vectorizadas, broadcasting, algebra lineal, y funciones matematias en array. Es la base sobre la que se basan muchas librerias en Python.

Nota: El arreglo de Numpy no es el mismo que la lista de pyhton, este no puede tener diferentes tipos de datos en el mismo.

```python
# pip install numpy
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) #array de 2 x 3
V = np.array([10,20,30]) #array 1 x 3
print(A + V)

prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba + prueba2)
```

```python
# Mas funciones de Numpy
z = np.zeros((3,3))
print(z)
o = np.ones((3,3))
print(o)
r = np.random.random((3,3))
print(r)
```

```python
# Rangos numericos
nums = np.arange(0,10,2) #saca los valores entre 0 y 10 de dos en dos
print(nums)

x = np.linspace(0,1,5) # saca cinco valores espaciados de manera equitativa entre el 0 y 1
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a + b)
print(a - b)
print(a / b)
print(a * b)
print(a.dot(b))
```

```python
c = np.array([[1,2,3],[2,4,6]])
d = np.array([[7,2,1],[3,4,7],[6,1,8]])
print(c.dot(d))
```

```python
data = np.array([1,2,3,4,5])
print(data.min())
print(data.max())
print(data.sum())
print(data.mean())
print(data.std())
```

```python
A = np.array([1,6,3,4,9])
print(A[A>5])
```
