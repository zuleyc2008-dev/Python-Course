# ESTRUCTURAS DE DATOS
Las estructuras de datos son formas de organizar información para trabajar con ellas de manera eficiente.

##LISTAS
Pueden ser ordenadas, modificadas y permiten duplicados.

## TUPLAS
Son ordenadas inmutables, no se pueden editar, y permiten duplicados

##SET
No son ordenadas, y no permiten duplicados.

##DICCIONARIOS
Son pares por clave y valor


NOTA:
Existen pilas, colas, filas, listas enlazadas o doblementeenlazadas. Pilas pertenece a como trabaja internamente, en este caso trabajo como (First in, last out) lo primero que entra es lo ultimo en salir. FILO
Las colas trabajan como una fila, el primero en formarse es el primero en salir. FIFO


## Listas
Son muy parecidas a los arreglos, pueden definir un espacio y almacenar todas las variables que tu quieras, y te puedes referir a ellas como tu quieras.
Son colecciones ordenadas que podemos modificiar, y almacenan elementos de distintos tipos de datos.

```python
Lista_de_compras = {'manzana','pan','leche','huevos','mantequilla','queso'}
print(Lista_de_compras)
```

```python
#Agregar un elemento a la lista, append()
Lista_de_compras.add('arroz')
print(Lista_de_compras)
```

```python
# 'Lista_de_compras' fue definida como un 'set' (conjunto), que no tiene el método 'insert()'.
# Para insertar un elemento en una posición específica, necesitamos una 'list' (lista).
# Convertimos el set a una lista, insertamos el elemento y luego actualizamos 'Lista_de_compras' como una lista.
temp_list = list(Lista_de_compras)
temp_list.insert(3, 'platano')
Lista_de_compras = temp_list # Ahora Lista_de_compras es una lista
print(Lista_de_compras)
```

```python
print(list(Lista_de_compras)[5])
```

```python
# Para eliminar un elemento de una lista existe el remove y el POP

Lista_de_compras.remove('arroz')
print(Lista_de_compras)
```

```python
Lista_de_compras.pop() #Sin nada dentro del parentesis elimina el ultimo elemento.
print(Lista_de_compras)
```

```python
Lista_de_compras.pop(4)
print(Lista_de_compras)
```

```python
# Podemos recorrer el arreglo por medio de un For
for elementos in Lista_de_compras:
    print(elementos)
```

```python
for i in range(len(Lista_de_compras)):           #len saca el tamaño de tu lista
    print(Lista_de_compras[i])
```

```python
print(len(Lista_de_compras))
for i in range(len(Lista_de_compras)):
    print(f'{i + 1} - {Lista_de_compras[i]}')
```

```python
promedios = [55,97.8,89.7,78.6,43,99.6]
promedios.sort()   #sirve para ordenar la lista de manera ascendente
print(promedios)
```

```python
#ordenar una lista de forma descendente
promedios.sort(reverse=True)
print(promedios)
```

```python
#Podemos ingresar al ultimo elemento de una lista
print(promedios[-1])
```

```python
#Encontrar el menor y mayor de una lista
print(max(promedios))
print(min(promedios))
```

```python
#Declaración y uso de matrices
calificaciones = [[100,88,93],[99,56,78],[98,76,54]]
i = 1
for cali_i in calificaciones:
    print(f'/nEstudiantes {i}')
    j = 1
    for cali_j in cali_i:
        print(f'cali{j}; {cali_j}')
```

```python
# Intersección
ca = {1244,1234,1265,1289,1456}
al = {1232,1244,1897,1094,1234}
#interseccion con .interseccion
ca_al = ca.intersection(al)
print(ca_al)
```

```python
#Diferencial
solo_calculo = ca.difference(al)
print(solo_calculo)
```

# DICCIONARIOS

```python
alumno = {
    'nombre': 'Daira Zuley',
    'edad': 17,
    'carrera': 'Ingenieria Aeronautica',
    'semestre': 1,
    'promedio': 98,
    'calificaciones': [98,99,100,100,100]
}
print(type(alumno))
alumno
```

```python
print(f"El alumno {alumno['nombre']} esta estudiando en el semestre {alumno['semestre']}, en la carrera de {alumno['carrera']}, tiene la edad de {alumno['edad']}, cuenta con un promedio de {alumno['promedio']}, sus calificaciones son {alumno['calificaciones']}")
```

```python
area_preferida = 'Química de los materiales'
alumno['area_preferida'] = area_preferida # Añadimos la nueva clave al diccionario
print(alumno['area_preferida'])
```

```python
for valor in alumno:
    print(alumno[valor])
#devuelve una vista de todos los valores y sus claves del diccionario
```

```python
# el metodo .clear() tambien se puede usar aqui
# tener en cuenta que el .get nos ayuda a obtener el valor (value) de cuerta llave (key)
print(alumno.get('nombre'))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
# El metodo .values() sirve para obtener los valores del diccionario tambien se puede usar en la interaccion
alumno_values = alumno.values()
print(alumno_values)
```

```python
if 'Carrera' in alumno:
    print('existe carrera')
else:
    print('no existe carrera')

if('dinero') in alumno:
    print('alumno tiene dinero')
else:
    print('el alumno no tiene dinero')
```
