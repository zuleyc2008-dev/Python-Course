# Operaciones Matemáticas

```
# 4to nivel de importancia
#Suma Ordinaria
suma = 3 + 4
variable1, variable2 = 3, 4
suma = variable1 + variable2
print(suma)

#Resta Ordinaria
variable1, variable2 = 15, 90
resta = variable1 - variable2
print(resta)
```

```
# 3er nivel de Importancia
# Multiplicacion Ordinaria
print('Multiplicación Ordinaria')
variable1, variable2 = 15, 78
multiplicación = variable1 * variable2
print(multiplicación)

#División Ordinaria
print('División Ordinaria')
variable1, variable2 = 78, 15
división = variable1 / variable2
print(división)


#División de Piso
print('División de Piso')
variable1, variable2 = 78, 15
división_de_piso = variable1 // variable2
print(división_de_piso)

#Potencias
print('Potencias')
variable1, variable2 = 2, 3
potencia = variable1 ** variable2
print(potencia)
```
